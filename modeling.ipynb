```python
import pandas as pd
import os
import pickle
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv(os.path.join("data","cars.csv"))
data
```

```python
features = ['AWD','RWD','Engine Size (l)','Cyl','Horsepower(HP)','City Miles Per Gallon','Highway Miles Per Gallon','Weight','Wheel Base','Len','Width']
feature_data = data[features]
type_dummy = pd.get_dummies(data['Type']).astype(int)
feature_data = pd.concat([feature_data,type_dummy],axis=1)
feature_data
```

```python
X = feature_data.to_numpy()
y = data['Retail Price'].to_numpy()
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
with open('model.pkl','wb') as f:
    pickle.dump(model,f)
```

```python
with open('model.pkl', 'rb') as f:
    model2 = pickle.load(f)
```
